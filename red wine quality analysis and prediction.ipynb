### Libraries Import

```
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn
```

### Data Aquisition

```
df_wine=pd.read_csv('/work/winequality-red.csv')
df_wine
```

### Shape Of Dataset

```
df_wine.shape
```

### First 5 Rows

```
df_wine.head()
```

### Last 5 rows

```
df_wine.tail()
```

### Data Types Of Features

```
df_wine.dtypes
```

### Convert Quality Feature From Integer To Float

```
df_wine['quality']=df_wine['quality'].astype('float64')
```

```
df_wine.dtypes
```

```
df_wine[['quality']].head()
```

### Quality Valuecounts

```
df_wine[['quality']].value_counts()
```

### Memory Usage

```
df_wine.memory_usage()
```

### Descriptive Statistics

```
df_wine.describe(include='all')
```

### Columns Name

```
df_wine.columns
```

### Taking x as Predictor And Independent Variable And Storing Features [['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar','chlorides', 'free sulfur dioxide', 'total sulfur dioxide', 'density','pH', 'sulphates', 'alcohol']] In It To Predict Quality

```
x=df_wine[['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar',
       'chlorides', 'free sulfur dioxide', 'total sulfur dioxide', 'density',
       'pH', 'sulphates', 'alcohol']]
```

### Taking y as Target And Dependent Variable And Storing Feature [['Quality']] In It

```
y=df_wine[['quality']]
```

### Train-Test Split For M.L.

```
from sklearn.model_selection import train_test_split
```

### Spliting Dataset 80% For Training And 20% For Testing 

```
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=0,test_size=0.2)
```

### Shape Of Train And Test Dataset

```
print(x_train.shape,y_train.shape)
print(x_test.shape,y_test.shape)
```

### By K-Nearest Neighbors(KNN)

```
from sklearn.neighbors import KNeighborsClassifier
```

### Fitting Of KNN

```
KNN=KNeighborsClassifier(n_neighbors=4)
KNN.fit(x_train,y_train)
```

### Predicting Quality By KNN

```
y_pred_KNN=KNN.predict(x_test)
```

### Predicted Quality By KNN

```
y_pred_KNN
```

### Accuracy Score Import

```
from sklearn.metrics import accuracy_score
```

### Accuracy Score For KNN

```
y_pred_KNN_accuracy=accuracy_score(y_test,y_pred_KNN)
y_pred_KNN_accuracy
```

### By Decision Tree Classifier(DTC)

```
from sklearn.tree import DecisionTreeClassifier
```

### Fitting Of DTC

```
DTC=DecisionTreeClassifier(criterion='entropy',max_depth=5)
DTC.fit(x_train,y_train)
```

### Predicting Quality By DTC

```
y_pred_DTC=DTC.predict(x_test)
```

### Predicted Quality By DTC

```
y_pred_DTC
```

### Accuracy Score For DTC

```
y_pred_DTC_accuracy=accuracy_score(y_test,y_pred_DTC)
y_pred_DTC_accuracy
```

### By Random Forest Classifier(RFC)

```
from sklearn.ensemble import RandomForestClassifier
```

### Fitting Of RFC

```
RFC=RandomForestClassifier(max_depth=11)
RFC.fit(x_train,y_train)
```

### Predicting Quality By RFC

```
y_pred_RFC=RFC.predict(x_test)
```

### Predicted Quality By RFC

```
y_pred_RFC
```

### Accuracy Score For RFC

```
y_pred_RFC_accuracy=accuracy_score(y_test,y_pred_RFC)
y_pred_RFC_accuracy
```

### By Gaussian Naive Bayes(GNB)

```
from sklearn.naive_bayes import GaussianNB
```

### Fitting Of GNB

```
GNB=GaussianNB()
GNB.fit(x_train,y_train)
```

### Predicting Quality By GNB

```
y_pred_GNB=GNB.predict(x_test)
```

### Predicted Quality By GNN

```
y_pred_GNB
```

### Accuracy Score For GNB

```
y_pred_GNB_accuracy=accuracy_score(y_test,y_pred_GNB)
y_pred_GNB_accuracy
```

### Outcome Of Analysis And Prediction

#### Random Forest Classifier is the better Classification model to predict quality of red wine because it's accuracy score is 73%(approx) i.e. the actual and predicted test values are very much similar to each other..
